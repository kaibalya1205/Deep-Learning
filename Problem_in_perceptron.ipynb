```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data=pd.DataFrame()
and_data=pd.DataFrame()
xor_data=pd.DataFrame()
```

```python
or_data['input_1']=[0,0,1,1]
or_data['input_2']=[0,1,0,1]
or_data['output']=[0,1,1,1]
```

```python
and_data['input_1']=[0,0,1,1]
and_data['input_2']=[0,1,0,1]
and_data['output']=[0,0,0,1]
```

```python
xor_data['input_1']=[0,0,1,1]
xor_data['input_2']=[0,1,0,1]
xor_data['output']=[0,1,1,0]
```

```python
or_data
```

```python
sns.scatterplot(x=or_data['input_1'],y=or_data['input_2'],hue=or_data['output'],s=200)
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('OR Gate Output Visualization')
plt.show()
```

```python
and_data
```

```python
sns.scatterplot(x=and_data['input_1'],y=and_data['input_2'],hue=and_data['output'],s=200)
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('AND Gate Output Visualization')
plt.show()
```

```python
xor_data
```

```python
sns.scatterplot(x=xor_data['input_1'],y=xor_data['input_2'],hue=xor_data['output'],s=200)
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('XOR Gate Output Visualization')
plt.show()
```

```python
from sklearn.linear_model import Perceptron
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf2.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
clf1.coef_
```

```python
clf1.intercept_
```
