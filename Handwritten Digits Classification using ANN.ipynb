```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
# Scaling  as minimum value of pixel is 0 and maximum is 255 we want to normalize from 0 to 1.
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
# NN Model creation
model=Sequential()
model.add(Flatten(input_shape=(28,28)))  # convert 2D array to 1D array 
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax')) # for more than one output nodes always use softmax
```

```python
model.summary()
```

```python
# Compile model
#categorical_crossentropy->label need onhot encoding
#sparse_categorical_crossentropy->label do not need onehot encoding
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
model.predict(X_test)  # gives probability of 0,1,2,3.....9 for every image i.e 10,000
```

```python
y_probability=model.predict(X_test)
```

```python
y_probability.argmax(axis=1) # we want for every array what is the maximum probability 
```

```python
y_pred=y_probability.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
# To improve accuracy, we can add more hidden layers
# increase epochs ex. 25
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])  # this shows training loss gradually decreaseing but validation loss after some time remain
#
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
# training data accuracy almost reaches 100 percent, but validation accuracy only reaches 97, arround indicating overfitting is happening
```
