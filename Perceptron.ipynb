```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
X = df.iloc[:, 0:2]   # takes all rows, first two columns as features
y = df.iloc[:, -1]    # takes all rows, last column as target
```

```python
p.fit(X,y)
```

```python
p.coef_ # weights for cgpa and resume
```

```python
p.intercept_ #bias
```

```python
pip install mlxtend
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X.values, y=y.values, clf=p, legend=2)
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.title('Perceptron Decision Boundary')
```
